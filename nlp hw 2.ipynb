```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
wor = tokenizer.tokenize(txt)
text = ' '.join(wor)
print(text)
```

теперь тестирую каждый теггер изапписываю результаты в список спписков (с ним легче работать, чем со списком словарей)

```python
from pymorphy2.tokenizers import simple_word_tokenize
pym_a = []
pym = []
for token in simple_word_tokenize(text):
    p = m.parse(token)[0]
    pym_a.append(p)
for item in pym_a:
    new_list = [item.word, item.tag.POS]
    pym.append(new_list)
print(pym)
```

```python
spc = []
doc = nlp(text)
for token in doc:
    if token.pos_ != 'SPACE':
        spc.append([token.text, token.pos_])
print(spc)
```

```python
mys = []
ana = p.analyze(text)
for word in ana:
    if 'analysis' in word:
        try: #один из элементов списка почему-то не читается - пишет, что нет [0], это решение пропустить плохоое, но я соовершенно не понимаю, как это исправить :(
            gr = word['analysis'][0]['gr']
            pos = gr.split('=')[0].split(',')[0]
        except:
            pos = 'не получилось'
        mys.append([word['text'], pos])
print(mys)
```

мои теги лежат в csv-файле, поэтому читаю его и превращаю в такой же список списков. проверяю длину, потому что буду использвать ее при подсчете accuracy ну и вооообще на всякий случай

```python
import csv
hands = []
with open('/Users/kseniadudiy/Downloads/tagsnlp.csv') as f:
    reader = csv.reader(f)
    for line in reader:
        z = [line[0], str(line[1:])]
        hands.append(z)
print(len(hands))
```

в моем золотом стандарте теги хотя бы частично совпадают с теми, которые предложены в трех теггерах (в некоторых ячейках у меня по два тега, чтобы совпадали с разными теггерами). поэтому сразу начинаю сравнение: сначала поиском по началу строки ищу совпадающиеслова, потом проверяю, совпадает ли тег с однимиз предложенных в золотом стандарте или его частью (в случае, например CCONJ и CONJ - в моей таблице CONJ, который совпадает и с одинаковым ему и с CCONJ). если есть совпадение, добавляю в отдельныйсписок. потом длину полученного списка, которая равна кол-ву совпадений, использую для подсчета accuracy. 

```python
import re
mys_c = []
for item in hands:
    for element in mys:
        if re.match(item[0], element[0]) != None:
            if re.search(item[1], element[1]) != None:
                mys_c.append(str(element))
                mys_c = list(dict.fromkeys(mys_c))
mystem_acc = len(mys_c)/len(hands)
print(mystem_acc)
```
